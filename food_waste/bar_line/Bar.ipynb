```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("gidaisrafi.csv")

df["loss_quantity"] = pd.to_numeric(df["loss_quantity"], errors="coerce")

df = df.dropna(subset=["loss_quantity"])

ulke_ceviri = {
    "Mexico": "Meksika",
    "Pakistan": "Pakistan",
    "Azerbaijan": "Azerbaycan",
    "Ghana": "Gana",
    "Thailand": "Tayland"
}

df["country"] = df["country"].replace(ulke_ceviri)

toplam_israf = df.groupby("country")["loss_quantity"].sum()

en_cok_israf = toplam_israf.sort_values(ascending=False).head(5)

plt.figure(figsize=(6,4))
en_cok_israf.plot(kind='bar', color='dodgerblue')
plt.title("En Çok Gıda İsrafı Yapan 5 Ülke")
plt.xlabel("Ülke")
plt.ylabel("Toplam Gıda İsrafı (ton)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
