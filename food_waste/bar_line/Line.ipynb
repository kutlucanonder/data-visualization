```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("gidaisrafi.csv")

df["loss_percentage"] = pd.to_numeric(df["loss_percentage"], errors="coerce")
df["year"] = pd.to_numeric(df["year"], errors="coerce")

df = df.dropna(subset=["loss_percentage", "year"])

df = df[df["year"] != 2022]

yillik_ortalama_israf = df.groupby("year")["loss_percentage"].mean()

plt.figure(figsize=(10,6))
plt.plot(
    yillik_ortalama_israf.index,
    yillik_ortalama_israf.values,
    marker='o', linestyle='-', color='dodgerblue'
)
plt.title("Yllara Gre Ortalama Gda sraf Oran")
plt.xlabel("Yl")
plt.ylabel("sraf Oran (%)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
